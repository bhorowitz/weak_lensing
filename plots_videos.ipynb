```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
delta_x = np.loadtxt('build/delta_x_32.txt')
plt.plot(delta_x[:,1], 'g', lw = 2, label = 'original')
plt.plot(delta_x[:,2], 'b', lw = 2, label = 'lognormal')
plt.plot(delta_x[:,3], 'r', lw = 2, label = '+noise')
plt.plot(delta_x[:,4], 'k', lw = 2, label = 'reconstructed')
plt.axis([0,32,-3,10])
plt.legend(loc = 2)
plt.savefig('lbfgs_32.eps')
```

```python
from matplotlib import animation

fig = plt.figure()
ax = plt.axes(xlim = (0, 32), ylim = (-3, 4))
line, = ax.plot([], [], 'k', lw = 2)
line1, = ax.plot([], [], 'g', lw = 2)
iter_text = ax.text(0.05, 0.85, '', transform=ax.transAxes)
iter_text.set_fontsize(15)

def init():
    #im = plt.plot(delta_x[:,1], 'g', iters[0], 'k')
    #return im,
    line.set_data([], [])
    line1.set_data([], [])
    iter_text.set_text('')
    return line, line1, iter_text

def animate(i):
    #im = plt.plot(delta_x[:,1], 'g', iters[i], 'k')
    #return im,
    line.set_data(delta_x[:,0], iters[i][2:])
    line1.set_data(delta_x[:,0], delta_x[:,1])
    iter_text.set_text('Iteration: ' + str(i) + ", $-2\ln(\pi)$: "  + str(iters[i][0]))
    return line, line1, iter_text


anim = animation.FuncAnimation(fig, animate, init_func = init, frames = iters.shape[0], interval = 10, blit = True)
anim.save('lbfgs_iters_32.mp4', fps = 5)
```
